```python
import csv
import random
import matplotlib
```

```python
n=int(input("Veuillez donner le nombre de produits:"))
with open("ventes.csv", mode="w", newline="") as fichier:
    writer = csv.writer(fichier)

    writer.writerow(["ID", "Prix", "Quantite", "Remise"])

    for i in range(1, n+1):
        prix = round(random.uniform(10, 500), 2)
        quantite = random.randint(1, 10)
        remise = round(random.uniform(0, 0.3), 2)
        writer.writerow([i, prix, quantite, remise])

print("Fichier ventes.csv généré avec succès !")
print("Il contient",n,"lignes + 1 ligne d'en-tête.")
print("Colonnes : ID, Prix, Quantite, Remise")
```

```python
rows = []

with open("ventes.csv", mode="r") as fichier:
    reader = csv.DictReader(fichier)

    for ligne in reader:
        prix = float(ligne["Prix"])
        quantite = int(ligne["Quantite"])

        ca_brut = round(prix * quantite, 2)

        ligne["CA_Brut"] = ca_brut
        rows.append(ligne)

with open("etape2_ca_brut.csv", mode="w", newline="") as fichier:
    fieldnames = ["ID", "Prix", "Quantite", "Remise", "CA_Brut"]
    writer = csv.DictWriter(fichier, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(rows)

print("etape2_ca_brut.csv créé avec succès !")
```

```python
rows = []

with open("etape2_ca_brut.csv", mode="r") as fichier:
    reader = csv.DictReader(fichier)

    for ligne in reader:
        ca_brut = float(ligne["CA_Brut"])
        remise = float(ligne["Remise"])

        ca_net = round(ca_brut * (1 - remise), 2)

        ligne["CA_Net"] = ca_net
        rows.append(ligne)

with open("etape3_ca_net.csv", mode="w", newline="") as fichier:
    fieldnames = ["ID", "Prix", "Quantite", "Remise", "CA_Brut", "CA_Net"]
    writer = csv.DictWriter(fichier, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(rows)

print("etape3_ca_net.csv créé avec succès !")
```

```python
meilleur_id = None
meilleur_ca_net = 0

with open("etape3_ca_net.csv", mode="r") as fichier:
    reader = csv.DictReader(fichier)
    for ligne in reader:
        ca_net = float(ligne["CA_Net"])
        if ca_net > meilleur_ca_net:
            meilleur_ca_net = ca_net
            meilleur_id = ligne["ID"]

print("Produit avec le plus gros bénéfice : ID", meilleur_id)
print("CA Net :", meilleur_ca_net, "DT")
```

```python
rows = []

with open("etape3_ca_net.csv", mode="r") as fichier:
    reader = csv.DictReader(fichier)

    for ligne in reader:
        ca_net = float(ligne["CA_Net"])

        tva = round(ca_net * 0.20, 2)

        ligne["TVA"] = tva
        rows.append(ligne)

with open("resultat_final.csv", mode="w", newline="") as fichier:
    fieldnames = ["ID", "Prix", "Quantite", "Remise", "CA_Brut", "CA_Net", "TVA"]
    writer = csv.DictWriter(fichier, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(rows)

print("resultat_final.csv créé avec succès !")
```

```python
ca_total = 0

with open("resultat_final.csv", mode="r") as fichier:
    reader = csv.DictReader(fichier)

    for ligne in reader:
        ca_total += float(ligne["CA_Net"])

ca_total = round(ca_total, 2)
print(f"  CA TOTAL DE L'ENTREPRISE : {ca_total:.2f} €")
```

```python
matplotlib.use("Agg")
import matplotlib.pyplot as plt

# Lire les données depuis ventes.csv
ids = []
ca_brut = []
ca_net = []
tva = []

with open("resultat_final.csv", mode="r") as fichier:
    reader = csv.DictReader(fichier)
    for ligne in reader:
        ids.append(str(ligne["ID"]))
        ca_brut.append(float(ligne["CA_Brut"]))
        ca_net.append(float(ligne["CA_Net"]))
        tva.append(float(ligne["TVA"]))

# Graphique 1 : CA Brut vs CA Net
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle("Analyse des Ventes", fontsize=14, fontweight="bold")

x = list(range(len(ids)))
largeur = 0.35

for i in range(len(x)):
    axes[0].bar(
        x[i] - largeur / 2,
        ca_brut[i],
        width=largeur,
        color="#4C72B0",
        label="CA Brut" if i == 0 else "",
    )
    axes[0].bar(
        x[i] + largeur / 2,
        ca_net[i],
        width=largeur,
        color="#55A868",
        label="CA Net" if i == 0 else "",
    )

axes[0].set_title("CA Brut vs CA Net par Produit")
axes[0].set_xlabel("ID Produit")
axes[0].set_ylabel("Montant (DT)")
axes[0].set_xticks(x)
axes[0].set_xticklabels(ids)
axes[0].legend()

# Graphique 2 : TVA par produit
axes[1].bar(ids, tva, color="#C44E52")
axes[1].set_title("Montant TVA par Produit")
axes[1].set_xlabel("ID Produit")
axes[1].set_ylabel("TVA (DT)")

plt.tight_layout()
plt.savefig("graphiques_ventes.png", dpi=150)
plt.show()
print("Graphiques générés avec succès !")
```
